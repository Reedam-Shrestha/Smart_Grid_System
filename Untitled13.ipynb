```python
import time
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, IntegerType, DoubleType, StringType
from pyspark.sql import functions as F # CORRECTED: Removed extra 's' from pyspark

# --- Data Generation (Fallback if CSVs not found) ---
def generate_sample_data(num_nodes=100, num_sources=5, hours_in_cycle=24):
    """Generates sample demand, source, and topology data for testing."""
    print("Generating sample data...")
    # Demand Data
    demand_data = []
    for node_id in range(1, num_nodes + 1):
        for hour in range(1, hours_in_cycle + 1):
            demand_kw = random.randint(80, 200)
            demand_data.append({'node_id': node_id, 'hour': hour, 'demand_kw': demand_kw})
    demand_df = pd.DataFrame(demand_data)

    # Source Capacity Data
    source_data = []
    source_types = ['Solar', 'Wind', 'Grid']
    for source_id in range(1, num_sources + 1):
        for hour in range(1, hours_in_cycle + 1):
            source_type = random.choice(source_types)
            capacity_kw = random.randint(100, 800)
            cost_per_kw = round(random.uniform(0.05, 0.15), 2)
            source_data.append({
                'source_id': source_id,
                'source_type': source_type,
                'hour': hour,
                'capacity_kw': capacity_kw,
                'cost_per_kw': cost_per_kw
            })
    source_df = pd.DataFrame(source_data)

    # Network Topology Data
    topology_data = []
    for source_id in range(1, num_sources + 1):
        for node_id in range(1, num_nodes + 1):
            distance_km = random.randint(1, 20)
            loss_coefficient = round(random.uniform(0.001, 0.02), 3)
            topology_data.append({
                'source_id': source_id,
                'node_id': node_id,
                'distance_km': distance_km,
                'loss_coefficient': loss_coefficient
            })
    topology_df = pd.DataFrame(topology_data)

    return demand_df, source_df, topology_df

# --- Chromosome Utilities ---
def initialize_population(pop_size, gene_pool, chromosome_length):
    """
    Initializes a population of chromosomes.
    Each chromosome is a list where each element (gene) is a source_id
    assigned to a corresponding demand point.
    """
    return [np.random.choice(gene_pool, size=chromosome_length).tolist() for _ in range(pop_size)]

def crossover(parent1, parent2, chromosome_length):
    """
    Performs single-point crossover between two parents to produce one child.
    """
    if chromosome_length < 2:
        return random.choice([parent1, parent2])

    point = random.randint(1, chromosome_length - 1)
    child = parent1[:point] + parent2[point:]
    return child

def mutate(chromosome, gene_pool, mutation_rate):
    """
    Applies mutation to a chromosome.
    For each gene, there's a chance (mutation_rate) for it to be replaced
    with a random gene from the gene_pool.
    """
    mutated_chromosome = list(chromosome) # Create a mutable copy
    for i in range(len(mutated_chromosome)):
        if random.random() < mutation_rate:
            mutated_chromosome[i] = random.choice(gene_pool)
    return mutated_chromosome

# --- Parallel Fitness Evaluation (Modified for PySpark) ---
def evaluate_population_parallel_spark(population, spark_demand_df, spark_source_df, spark_topology_df):
    """
    Evaluates the fitness of an entire population using PySpark.
    """
    # Use the existing SparkSession rather than creating a new one each time in this function
    spark = SparkSession.builder.appName("FitnessEvaluator").getOrCreate()

    print("--- Fitness Evaluation: Converting Spark DFs to Pandas for Broadcast ---")
    source_grouped_pd = spark_source_df.groupBy("source_id", "hour").agg(
        F.mean("capacity_kw").alias("capacity_kw"),
        F.mean("cost_per_kw").alias("cost_per_kw")
    ).toPandas().set_index(['source_id', 'hour'])
    print("--- Fitness Evaluation: source_grouped_pd created ---")

    topology_lookup_pd = spark_topology_df.toPandas().set_index(['source_id', 'node_id'])
    print("--- Fitness Evaluation: topology_lookup_pd created ---")

    demand_pd = spark_demand_df.toPandas()
    print("--- Fitness Evaluation: demand_pd created ---")

    # Broadcast these pandas DataFrames
    b_demand_pd = spark.sparkContext.broadcast(demand_pd)
    b_source_grouped_pd = spark.sparkContext.broadcast(source_grouped_pd)
    b_topology_lookup_pd = spark.sparkContext.broadcast(topology_lookup_pd)
    print("--- Fitness Evaluation: DataFrames broadcasted ---")

    # Define the function that calculates fitness for a single chromosome.
    # This function will run on Spark executors.
    def calculate_single_chromosome_fitness(chromosome_as_list):
        """
        Calculates fitness for a single chromosome using broadcasted Pandas DataFrames.
        This function runs on a Spark worker.
        """
        # Retrieve broadcasted variables
        demand_data = b_demand_pd.value
        source_grouped = b_source_grouped_pd.value
        topology_lookup = b_topology_lookup_pd.value

        total_cost = 0
        penalty = 0

        for idx, gene_source_id in enumerate(chromosome_as_list):
            # Print every 1000 iterations to avoid flooding output, adjust based on your chromosome length
            if idx % 1000 == 0 and idx > 0: # Added idx > 0 to avoid printing for the very first item consistently
                print(f"  > Processing demand point {idx}/{len(chromosome_as_list)} for current chromosome...")
            try:
                # Get demand details from the broadcasted demand_data
                demand_point = demand_data.iloc[idx]
                node_id = demand_point['node_id']
                hour = demand_point['hour']
                demand_kw = demand_point['demand_kw']

                # Lookup source capacity and cost from broadcasted source_grouped
                # Use .loc with a tuple for multi-index lookup
                capacity_info = source_grouped.loc[(gene_source_id, hour)]
                available_capacity = capacity_info['capacity_kw']
                cost_per_kw = capacity_info['cost_per_kw']

                # Lookup transmission distance and loss coefficient from broadcasted topology_lookup
                # Use .loc with a tuple for multi-index lookup
                connection_info = topology_lookup.loc[(gene_source_id, node_id)]
                distance = connection_info['distance_km']
                loss_coefficient = connection_info['loss_coefficient']

                # Calculate energy needed at source, considering transmission loss
                energy_at_source_needed = demand_kw * (1 + loss_coefficient * distance)

                # Constraint: Max Capacity - Penalize if source cannot meet adjusted demand
                if available_capacity < energy_at_source_needed:
                    penalty += 1000 * (energy_at_source_needed - available_capacity)

                # Calculate cost for this specific allocation
                total_cost += energy_at_source_needed * cost_per_kw

            except KeyError:
                # This occurs if a (source_id, hour) or (source_id, node_id) mapping is missing
                penalty += 50000
            except Exception as e:
                penalty += 100000

        return total_cost + penalty

    print("--- Fitness Evaluation: Starting parallel fitness calculation ---")
    fitness_scores = spark.sparkContext.parallelize(population).map(calculate_single_chromosome_fitness).collect()
    print("--- Fitness Evaluation: Parallel fitness calculation complete ---")

    return fitness_scores

# --- Tournament Selection ---
def tournament_selection(population, fitness_scores, k=3):
    """
    Selects individuals for the next generation using tournament selection.
    k specifies the number of individuals participating in each tournament.
    """
    selected = []
    population_with_fitness = list(zip(population, fitness_scores))
    for _ in range(len(population)):
        contenders = random.sample(population_with_fitness, k)
        winner = min(contenders, key=lambda x: x[1])[0]
        selected.append(winner)
    return selected

# --- Main Evolution Loop ---
def genetic_algorithm_spark(spark_demand_df, spark_source_df, spark_topology_df, num_generations, population_size, mutation_rate):
    """
    Main Genetic Algorithm loop implemented with PySpark for parallel fitness evaluation.
    """
    source_ids = [row.source_id for row in spark_source_df.select("source_id").distinct().collect()]
    chromosome_length = spark_demand_df.count()

    # Initial population
    population = initialize_population(population_size, source_ids, chromosome_length)

    best_fitness_history = []
    avg_fitness_history = []

    print("\n--- Starting Genetic Algorithm Evolution ---")
    for generation in range(num_generations):
        # Evaluate fitness
        fitness_scores = evaluate_population_parallel_spark(population, spark_demand_df, spark_source_df, spark_topology_df)

        # Track best and average fitness
        current_best_fitness = min(fitness_scores)
        current_avg_fitness = sum(fitness_scores) / len(fitness_scores)
        best_fitness_history.append(current_best_fitness)
        avg_fitness_history.append(current_avg_fitness)

        # Find the best solution in the current generation
        best_solution_idx = np.argmin(fitness_scores)
        best_solution_chromosome = population[best_solution_idx]

        print(f"Generation {generation+1}/{num_generations}: Best Fitness = {current_best_fitness:.2f}, Average Fitness = {current_avg_fitness:.2f}")

        # Selection
        selected_population = tournament_selection(population, fitness_scores)

        # Create next generation
        next_population = []
        # Keep the best solution to ensure elitism
        next_population.append(best_solution_chromosome)
        for i in range(population_size - 1): # -1 because we already added the best
            parent1 = random.choice(selected_population)
            parent2 = random.choice(selected_population)
            child = crossover(parent1, parent2, chromosome_length)
            mutated_child = mutate(child, source_ids, mutation_rate)
            next_population.append(mutated_child)

        population = next_population

    final_best_fitness = min(best_fitness_history)
    final_best_solution = population[np.argmin(fitness_scores)] # Get the best solution from the final population

    print("\n--- Genetic Algorithm Evolution Complete ---")
    print(f"Final Best Fitness: {final_best_fitness:.2f}")
    return final_best_solution, best_fitness_history, avg_fitness_history


# --- Baseline Method (Greedy Approach - Modified for PySpark) ---
def greedy_algorithm_spark(spark_demand_df, spark_source_df, spark_topology_df):
    """
    A simple greedy algorithm for energy distribution.
    For each demand point, it tries to assign the cheapest available source
    that can meet the demand (considering losses), prioritizing cost.
    This version converts Spark DataFrames to Pandas for calculation due to
    the iterative nature of the original greedy algorithm, which is less
    efficient to implement purely in Spark DataFrame operations without
    significant re-architecting (e.g., cross-joins and window functions).
    """
    # Use the existing SparkSession rather than creating a new one each time in this function
    spark = SparkSession.builder.appName("GreedySolver").getOrCreate()

    print("\n--- Running Greedy Algorithm (Baseline with PySpark Data converted to Pandas) ---")
    start_time = time.time()

    # Convert Spark DataFrames to Pandas DataFrames for the greedy algorithm logic.
    # This is done for simplicity to adapt the existing iterative greedy approach.
    # For very large datasets that don't fit in memory, a purely Spark-native greedy
    # algorithm would be required, which would involve different DataFrame operations.
    demand_pd = spark_demand_df.toPandas()
    source_grouped_pd = spark_source_df.groupBy(['source_id', 'hour']).agg(
        F.mean('capacity_kw').alias('capacity_kw'),
        F.mean('cost_per_kw').alias('cost_per_kw')
    ).toPandas().set_index(['source_id', 'hour'])
    topology_lookup_pd = spark_topology_df.toPandas().set_index(['source_id', 'node_id'])
    all_source_ids = [row.source_id for row in spark_source_df.select("source_id").distinct().collect()]

    total_cost = 0
    total_penalty = 0
    assigned_sources = [None] * len(demand_pd) # To store the greedy solution chromosome

    for idx, demand_point in demand_pd.iterrows():
        node_id = demand_point['node_id']
        hour = demand_point['hour']
        demand_kw = demand_point['demand_kw']

        best_assignment_cost = float('inf')
        best_source = None

        # Iterate through all possible sources for this demand point
        for source_id in all_source_ids:
            try:
                # Get source info
                capacity_info = source_grouped_pd.loc[(source_id, hour)]
                available_capacity = capacity_info['capacity_kw']
                cost_per_kw = capacity_info['cost_per_kw']

                # Get topology info
                connection_info = topology_lookup_pd.loc[(source_id, node_id)]
                distance = connection_info['distance_km']
                loss_coefficient = connection_info['loss_coefficient']

                energy_at_source_needed = demand_kw * (1 + loss_coefficient * distance)

                if available_capacity >= energy_at_source_needed:
                    current_assignment_cost = energy_at_source_needed * cost_per_kw
                    if current_assignment_cost < best_assignment_cost:
                        best_assignment_cost = current_assignment_cost
                        best_source = source_id
            except KeyError:
                # This source-node or source-hour combination does not exist, skip
                continue
            except Exception as e:
                continue

        if best_source is not None:
            total_cost += best_assignment_cost
            assigned_sources[idx] = best_source
        else:
            # If no source can meet the demand for this point, add a penalty
            total_penalty += 10000 # Large penalty for unmet demand
            assigned_sources[idx] = random.choice(all_source_ids) # Assign a random source to complete chromosome

    final_greedy_cost = total_cost + total_penalty
    end_time = time.time()
    greedy_runtime = end_time - start_time
    print(f"Greedy Algorithm Runtime (PySpark Data converted to Pandas): {greedy_runtime:.2f} seconds")
    print(f"Greedy Algorithm Total Cost: {final_greedy_cost:.2f}")
    print("-" * 30)
    return final_greedy_cost, assigned_sources, greedy_runtime

# --- Main Runner Function ---
def genetic_algorithm_spark_runner():
    # Initialize Spark Session
    # Ensure this is only created once.
    spark = SparkSession.builder \
        .appName("SmartGridOptimizationGA") \
        .config("spark.executor.memory", "2g") \
        .config("spark.driver.memory", "2g") \
        .config("spark.driver.maxResultSize", "2g") \
        .getOrCreate()

    print("--- Using Small Sample Data for Quick Execution ---")
    # Force generating very small sample data to ensure quick execution
    demand_df_pd, source_df_pd, topology_df_pd = generate_sample_data(num_nodes=5, num_sources=2, hours_in_cycle=3)
    spark_demand_df = spark.createDataFrame(demand_df_pd)
    spark_source_df = spark.createDataFrame(source_df_pd)
    spark_topology_df = spark.createDataFrame(topology_df_pd)

    print("--- Loaded Data Summary (PySpark DataFrames from Sample Data) ---")
    print("Demand Data:")
    spark_demand_df.printSchema()
    spark_demand_df.show(5)
    print(f"Total demand entries: {spark_demand_df.count()}") # This will now be very small

    print("\nSource Data:")
    spark_source_df.printSchema()
    spark_source_df.show(5)
    print(f"Total source entries: {spark_source_df.count()}")

    print("\nTopology Data:")
    spark_topology_df.printSchema()
    spark_topology_df.show(5)
    print(f"Total topology entries: {spark_topology_df.count()}")
    print("-" * 30)


    # --- Genetic Algorithm Parameters (TEMPORARILY REDUCED FOR FASTER TESTING) ---
    NUM_GENERATIONS = 5   # Reduced for quick test
    POPULATION_SIZE = 10  # Reduced for quick test
    CHROMOSOME_LENGTH = spark_demand_df.count() # Each gene maps a demand entry (row) to a source
    MUTATION_RATE = 0.05

    print(f"Genetic Algorithm Parameters (Current Test Run):")
    print(f"  Number of Generations: {NUM_GENERATIONS}")
    print(f"  Population Size: {POPULATION_SIZE}")
    print(f"  Chromosome Length (demand points): {CHROMOSOME_LENGTH}")
    print(f"  Mutation Rate: {MUTATION_RATE}")
    print("-" * 30)

    # Run Genetic Algorithm
    ga_start_time = time.time()
    best_solution_ga, best_fitness_history, avg_fitness_history = genetic_algorithm_spark(
        spark_demand_df, spark_source_df, spark_topology_df,
        NUM_GENERATIONS, POPULATION_SIZE, MUTATION_RATE
    )
    ga_end_time = time.time()
    ga_total_runtime = ga_end_time - ga_start_time

    # Run Greedy Algorithm (Baseline)
    greedy_cost, greedy_solution, greedy_runtime = greedy_algorithm_spark(
        spark_demand_df, spark_source_df, spark_topology_df
    )

    # --- Performance Analysis and Result Interpretation ---
    print("\n--- Performance Comparison ---")
    print(f"Genetic Algorithm Best Cost: {best_fitness_history[-1]:.2f}")
    print(f"Greedy Algorithm Total Cost: {greedy_cost:.2f}")

    if best_fitness_history[-1] < greedy_cost:
        print("Genetic Algorithm found a better solution than the Greedy Algorithm.")
    else:
        print("Greedy Algorithm found a comparable or better solution (or GA needs more generations/tuning).")

    print(f"\nRuntime Comparison:")
    print(f"Genetic Algorithm Runtime (PySpark): {ga_total_runtime:.2f} seconds")
    print(f"Greedy Algorithm Runtime (PySpark Data converted to Pandas): {greedy_runtime:.2f} seconds")
    print("-" * 30)

    # 2. Plot convergence curves
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, NUM_GENERATIONS + 1), best_fitness_history, label='Best Fitness')
    plt.plot(range(1, NUM_GENERATIONS + 1), avg_fitness_history, label='Average Fitness')
    plt.xlabel('Generation')
    plt.ylabel('Fitness Score (Cost)')
    plt.title('Genetic Algorithm Convergence Curve (PySpark)')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig('ga_convergence_pyspark.png') # Save plot instead of showing

    # Stop SparkSession at the end
    spark.stop()


# --- Run the Algorithms and Collect Data ---
if __name__ == "__main__":
    genetic_algorithm_spark_runner()
```
